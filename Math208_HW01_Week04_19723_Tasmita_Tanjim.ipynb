```python
#question 01

import random
import matplotlib.pyplot as plt

#declaring variable
random_num = []           #an empty list for storing the numbers
count = 0

#Generating  500 values within the range -20 to 20
while count < 500:                            #using a while loop  untill 500 so that the loop iterates 0 to 499
   random_num.append(random.uniform(-20,21))  #taking range 21 so that +20 is also included
   count = count + 1


#function of get the mean of the random numbers
def mean_of_nums(numbers):
  m = sum(numbers) / len(numbers)  #getting the mean by dividing the sum of numbers by the total amount of numbers
  return m


#function to get the median of the numbers
def median_of_nums(numbers):
  sorted_num = sorted(numbers)    #using sort function to get the numbers in increasing pattern
  length = len(sorted_num)       #calculating the length of the sorted_num

  if length % 2 == 0:                     #if the position is even
    med_1 = sorted_num[length // 2 - 1]   #calculating the position before the middle position
    med_2 = sorted_num[length // 2]
    med = (med_1 + med_2) / 2

  else:
     med =  sorted_num[length // 2]

  return med


#function to calculate the standard deviation
def std_deviation_of_nums(numbers):
    n = len(numbers)
    if n < 2:
        return 0.0  # Standard deviation is not possible for a single number

    mean = mean_of_nums(random_num)
    sum_sqr = 0.0

    for i in numbers:
      sum_sqr += (i - mean) ** 2

    std_dev = (sum_sqr / n) ** 0.5

    return std_dev



#Getting the results, mean, median and deviation
mean = mean_of_nums(random_num)
median =  median_of_nums(random_num)
std_deviation = std_deviation_of_nums(random_num)



#Printing all the results
print("Mean, Median and Standard deviation of all the random numbers between -20 to +20\n")
print("Mean : ", mean)
print("Median " , median)
print("Standard Deviation :", std_deviation,"\n")


# Plotting the histogram with 10 bins
plt.hist(random_num, bins=10, color='green', edgecolor='white')
plt.title("Histogram of Random Numbers")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()

```
